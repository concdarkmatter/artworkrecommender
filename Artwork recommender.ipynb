```python
def clean_data(x):
    if isinstance(x, list):
        return [str.lower(i.replace(" ", "")) for i in x]
    else:
        #Check if director exists. If not, return empty string
        if isinstance(x, str):
            return str.lower(x.replace(" ", ""))
        else:
            return ''
```

```python
features = ['artist', 'genre', 'style']

for feature in features:
    data[feature] = data[feature].apply(clean_data)
```

```python
data.head()
```

```python
data['bag_of_words'] = ''
columns = data.loc[:,data.columns!= 'title']
for index, row in data.iterrows():
    words = ''
    for col in columns:
        if col != 'artist':
            words = words + ''.join(row[col])+ ' '
        else:
            words = words + row[col]+ ' '
    row['bag_of_words'] = words
data.drop(columns = [col for col in data.loc[:,data.columns!= 'title'] if col!= 'bag_of_words'], inplace = True)
data.head()
```

```python
data.set_index('title', inplace = True)
data
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
```

```python
count = CountVectorizer(stop_words='english')
count_matrix = count.fit_transform(data['bag_of_words'])
print(count_matrix.shape)

indices = pd.Series(data.index)
indices[:5]
```

```python
cosine_sim = cosine_similarity(count_matrix, count_matrix)
cosine_sim
```

```python
def recommendations(title, cosine_sim = cosine_sim):
    
    recommended_movies = []
    
    idx = indices[indices == title].index[0]
    
    score_series = pd.Series(cosine_sim[idx]).sort_values(ascending = False)

    top_10_indexes = list(score_series.iloc[2:12].index)
    
    for i in top_10_indexes:
        recommended_movies.append(list(data.index)[i])
        
    return recommended_movies
```
